### Data Manipulation 

```python
#Converting "TotalCharges" to numeric
df_churn['TotalCharges']=pd.to_numeric(df_churn['TotalCharges'],errors="coerce")
df_churn.dtypes
```

```python
df_churn.isnull().sum()
```

```python
df_churn = df_churn.dropna() #Dropping the missing values in "TotalCharges"
```

```python
df_churn = df_churn.drop('customerID',axis=1) #Customer ids are unique and insignificant so it is better to drop this feature
```

```python
#replace values in columns with binary features

cols=['Partner','Dependents','PhoneService','PaperlessBilling']

for i in cols:
    df_churn[i]=df_churn[i].replace({"Yes":1,"No":0})
```

```python
#Encoding column 'gender'
df_churn['gender']=df_churn['gender'].replace({"Male":1,"Female":0})
```

```python
#transforming the data using Label encoder

df_churn = df_churn.apply(LabelEncoder().fit_transform)
df_churn.head(3)
```

### Implement Machine Learning Models

```python
#Separating the Target feature
X = df_churn.drop('Churn',axis = 1)
y = df_churn['Churn']
```

```python
print('Shape of X: {}'.format(X.shape))
print('Shape of y: {}'.format(y.shape))
```

```python
#Splitting the dataset for further analysis
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

#### Model 1 - Logistic Regression

```python
lr = LogisticRegression(penalty="l1",C=.1,solver='liblinear',random_state=0)
lr.fit(X_train,y_train)

y_pred_lr = lr.predict(X_test)

train_logistic = round((lr.score(X_train, y_train))*100,1)
test_logistic = round((lr.score(X_test, y_test))*100,1)

print("Training accuracy: %s %%" % train_logistic)
print('Test accuracy: %s %%' % test_logistic)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred_lr))

##Confusion matrix
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

confusion_matrix_logistic = metrics.confusion_matrix(y_pred_lr, y_test, [1,0])

plt.figure(figsize=(10,8))
sns.heatmap(confusion_matrix_logistic, annot=True, fmt='.2f',xticklabels = ["Left", "Stayed"] , yticklabels = ["Left", "Stayed"] ,cmap = plt.cm.Blues_r)
plt.ylabel('True class')
plt.xlabel('Predicted class')
plt.title('Logistic Regression')
```

##### Conclusion- Logistic 

* When I tried fitting the model on the test data using Logistic Regression, I got an accuracy of 80.8%.

* our model has around 80% Precision, Recall & F1 score on average which is pretty decent

# Model 2 - Random Forest 

```python
rfc = RandomForestClassifier(random_state=0,n_estimators=1000,criterion="gini")

rfc.fit(X_train,y_train)

y_pred_rfc = rfc.predict(X_test)

train_rfc = round((rfc.score(X_train, y_train))*100,1)
test_rfc = round((rfc.score(X_test, y_test))*100,1)

print("Training accuracy: %s %%" % train_rfc)
print('Test accuracy: %s %%' % test_rfc)
```

```python
print(classification_report(y_test, y_pred_rfc))

##Confusion matrix

confusion_matrix_rfc = metrics.confusion_matrix(y_pred_rfc, y_test, [1,0])

plt.figure(figsize=(10,8))
sns.heatmap(confusion_matrix_rfc, annot=True, fmt='.2f',xticklabels = ["Left", "Stayed"] , yticklabels = ["Left", "Stayed"] ,cmap = plt.cm.Blues_r)
plt.ylabel('True class')
plt.xlabel('Predicted class')
plt.title('RFC')
```

#### Random Forest Conclusion 

* Random Forest did not perform better than LOgistic Regression Model. We got around 79.3% accuracy

### Model 3 KNN  

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 3)

knn.fit(X_train,y_train)

y_pred_knn = knn.predict(X_test)

train_knn = round((knn.score(X_train, y_train))*100,1)
test_knn = round((knn.score(X_test, y_test))*100,1)

print("Training accuracy: %s %%" % train_knn)
print('Test accuracy: %s %%' % test_knn)
```

```python
print(classification_report(y_test, y_pred_knn))

##Confusion matrix

confusion_matrix_knn = metrics.confusion_matrix(y_pred_knn, y_test, [1,0])

plt.figure(figsize=(10,8))
sns.heatmap(confusion_matrix_knn, annot=True, fmt='.2f',xticklabels = ["Left", "Stayed"] , yticklabels = ["Left", "Stayed"] ,cmap = plt.cm.Blues_r)
plt.ylabel('True class')
plt.xlabel('Predicted class')
plt.title('kNN')
```

#### KNN Conclusion 

* KNN performed even worse than Random Forest by getting an accuracy of 75%.
